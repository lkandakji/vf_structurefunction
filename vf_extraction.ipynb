```r
library(readr)
library(tidyverse)
```

```r
file_path <- "/Users/lynnkandakji/Library/CloudStorage/OneDrive-UniversityCollegeLondon/Malena VF project/test_VF_new eyesuite.txt"
df <- read.table(file_path, sep = ";", header = FALSE, fill = TRUE)
colnames(df) <- as.character(unlist(df[1, ]))
df <- df[-1, ]
```

```r
start_col <- 53
num_cols <- ncol(df) - start_col + 1

# Initialize an empty vector for the new column names
new_col_names <- c()

# Populate the new_col_names vector with repeated patterns
for (i in 1:(num_cols / 5)) {
  new_col_names <- c(new_col_names, "XCord", "YCord", "Thres 1", "Thres 2", "Norm Value")
}

# Truncate the names to match the exact number of columns
new_col_names <- new_col_names[1:num_cols]

# Assign the new column names to the specific range in the data frame
names(df)[start_col:ncol(df)] <- new_col_names
```

```r
names(df)
print(head(df))
```

```r
row_index <- 2
thres1_values <- df[row_index, which(names(df) == "Thres 1")]

thres1_values <- as.numeric(thres1_values)
average_thres1 <- mean(thres1_values, na.rm = TRUE)
print(average_thres1)

thres1_filtered_values <- c()

for (i in seq(start_col, ncol(df), by = 5)) {
  xcord_value <- as.numeric(df[row_index, i])
  ycord_value <- as.numeric(df[row_index, i+1])
  thres1_value <- as.numeric(df[row_index, i+2])
  
  if (!is.na(xcord_value) && !is.na(ycord_value) && abs(xcord_value) <= 20 && abs(ycord_value) <= 20) {
    thres1_filtered_values <- c(thres1_filtered_values, thres1_value)
  }
}

average_thres1_filtered <- mean(thres1_filtered_values, na.rm = TRUE)
print(average_thres1_filtered)
```

```r
xc_columns <- df[, grep("XCord", names(df))]
xc_combined <- unlist(xc_columns, use.names = FALSE)
xc_range <- range(xc_combined, na.rm = TRUE)
print(xc_range)

yc_columns <- df[, grep("^YCord$", names(df))]
yc_columns <- lapply(yc_columns, as.numeric)
yc_combined <- unlist(yc_columns, use.names = FALSE)
yc_range <- range(yc_combined, na.rm = TRUE)
print(yc_range)
```

```r
df$MS_20 <- NA  # Create the new column

for (row_index in 1:nrow(df)) {
  thres1_filtered_values <- c()
  
  for (i in seq(start_col, ncol(df), by = 5)) {
    xcord_value <- as.numeric(df[row_index, i])
    ycord_value <- as.numeric(df[row_index, i+1])
    thres1_value <- as.numeric(df[row_index, i+2])
    
    if (!is.na(xcord_value) && !is.na(ycord_value) && abs(xcord_value) <= 20 && abs(ycord_value) <= 20) {
      thres1_filtered_values <- c(thres1_filtered_values, thres1_value)
    }
  }
  
  average_thres1_filtered <- mean(thres1_filtered_values, na.rm = TRUE)
  df$MS_20[row_index] <- average_thres1_filtered
}
```

```r
df$MS_30 <- NA  # Create the new column

for (row_index in 1:nrow(df)) {
  thres1_filtered_values <- c()
  
  for (i in seq(start_col, ncol(df), by = 5)) {
    xcord_value <- as.numeric(df[row_index, i])
    ycord_value <- as.numeric(df[row_index, i+1])
    thres1_value <- as.numeric(df[row_index, i+2])
    
    if (!is.na(xcord_value) && !is.na(ycord_value) && abs(xcord_value) <= 30 && abs(ycord_value) <= 30) {
      thres1_filtered_values <- c(thres1_filtered_values, thres1_value)
    }
  }
  
  average_thres1_filtered <- mean(thres1_filtered_values, na.rm = TRUE)
  df$MS_30[row_index] <- average_thres1_filtered
}
```
